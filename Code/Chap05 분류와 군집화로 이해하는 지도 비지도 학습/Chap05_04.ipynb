### 5.4 붓꽃 데이터를 이용한 k-NN 활용 연습


```python
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
iris.data[:5]     # 최초 5개 데이터의 값을 출력
```

```python
print('iris 데이터의 형태:', iris.data.shape)
print('iris 데이터의 속성들:', iris.feature_names)
print('iris 데이터의 레이블:', iris.target)

label_counts = np.bincount(iris.target)

for i, count in enumerate(label_counts):
   print(f'레이블 {i} : {count}개')
```

```python
import pandas as pd

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = pd.Series(iris.target)
iris_df.head()
```

```python
iris_df['target'].value_counts()
```

```python
iris_df.values
```

```python
X = iris_df.iloc[:, :4]
y = iris_df.iloc[:, -1]
print(X, y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

def iris_knn(X, y, k):
    X_train,X_test,y_train,y_test =train_test_split(X, y, test_size=0.3)
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    return metrics.accuracy_score(y_test, y_pred)

k = 3
scores = iris_knn(X, y, k)
print('n_neighbors가 {0:d}일때 정확도: {1:.3f}'.format(k, scores))
```

이번에는 훈련 데이터와 검증 데이터를 나누어서 훈련하지 않고 사용가능한 모든 데이터를 사용해서 모델을 학습해 보자. 이 때 n_neightbors 매개변수 값은 이전과 같이 3으로 주도록 하자.

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()
k = 3
knn = KNeighborsClassifier(n_neighbors = k)
knn.fit(iris.data, iris.target)
```

```python
classes = {0:'setosa', 1:'versicolor', 2:'virginica'}

# 새로운 데이터를 제시해 보자.
X = [[4, 2, 1.3, 0.4],
     [4, 3, 3.2, 2.2]]
y = knn.predict(X)

print('{} 특성을 가지는 품종: {}'.format(X[0], classes[y[0]]))
print('{} 특성을 가지는 품종: {}'.format(X[1], classes[y[1]]))
```
