### 4.5 데이터의 관계를 설명하는 선형 회귀 함수의 시각적 이해

```python
import numpy as np
from sklearn.metrics import mean_squared_error as mse

# 5개 점의 x, y 좌표값
x = np.array([1, 4.5, 9, 10, 13])
y = np.array([0, 0.2, 2.5, 5.4, 7.3])

w_list = np.arange(1.0, 0.2, -0.1)
for w in w_list:     # w를 바꾸어가며 예측치와 정답의 오차 비교
   y_hat = w * x
   print('w = {:.1f}, 평균제곱 오차: {:.2f}'.format(w, mse(y_hat, y)))
```

```python
# 그래프 그리기
import matplotlib.pyplot as plt

w_list = np.arange(1.0, 0.2, -0.1)
mse_list = []
for w in w_list:     # w를 바꾸어가며 예측치와 정답의 오차 비교
   y_hat = w * x
   mse_list.append(mse(y_hat, y))

plt.figure(figsize=(10, 6))
plt.plot(w_list, mse_list, 'o-', color='blue', linewidth=2, markersize=8, label='MSE(w)')

# 최저점(최소 오차) 강조
best_w = w_list[np.argmin(mse_list)]
best_mse = min(mse_list)
plt.plot(best_w, best_mse, 'ro', markersize=12, label=f'minima (w={best_w:.1f})')

plt.title('MSE(w)', fontsize=16)
plt.xlabel('weight w', fontsize=14)
plt.ylabel('error (MSE)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.xticks(w_list)  # x축에 w 값 정확히 표시
plt.tight_layout()
plt.show()
```
